```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("top_600_combined.csv")
url_path = ("https://imdb-api.com/en/API/SearchMovie/k_br0y2jj3/")
```

```python
df = pd.DataFrame(df)
```

```python
df.fillna(0, inplace=True)
```

## Last row was 391.
For next call, use df_func = df[491:591]

```python
start = 491
end = 591
```

```python
df_func = df[start:end]
```

```python
df_func
```

```python
for num, title in enumerate(df_func["Title"]):
    tit = df_func.loc[df_func["Movie_id"] == 0, "Title"]
    resp = requests.get(f"{url_path}{title}")
    resp_json1 = resp.json()
    id = resp_json1['results'][0]['id']
    image = resp_json1['results'][0]['image']
    print(id, title)
    df_func['Movie_id'][num:(num+1)] = id
    df_func['Image'][num:(num+1)] = image
```

```python
df_func.to_csv(f"{start}-{end}.csv")
```
